```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

import igraph
import os
import csv
import sys
import numpy as np
import pickle
from numpy.random import choice
from itertools import permutations
from random import shuffle
from tqdm import tqdm
import random

sys.path.insert(0, '../../')


from plot.plot_utils import *
```

```python
mapping_configurations = {
    "N1": {"s_m": 0.1, "h_m": 0.4, "h_M": 0.0},
    "N2": {"s_m": 0.45, "h_m": 0.5, "h_M": 0.0,},
    "N3": {"s_m": 0.3, "h_m": -0.25, "h_M": 0.0,},
    "N4": {"s_m": 0.3, "h_m": 0.5, "h_M": 0.2}
}

```

```python
#all_configurations = {}


for config in mapping_configurations:
    
    if config not in os.listdir("../../plot/sample_gephi/"):
        os.mkdir("../../plot/sample_gephi/" + config)
        
    
    s_m = mapping_configurations[config]["s_m"]

    hm = mapping_configurations[config]["h_m"]

    hM = mapping_configurations[config]["h_M"]
    
    
    graph = [x for x in os.listdir("../../semi_synthetic_graphs/data/synth/") if "s_m" + str(s_m) in x and "hm" + str(hm) in x and "hM" + str(hM) in x][0]
    
    with open("../../semi_synthetic_graphs/data/synth/" + graph, "rb") as f:
        bi_partite_graph = pickle.load(f)
        
    mapping_colors = mapping_colors_function(bi_partite_graph)
    
    indegree_nodes = {node.index: node.indegree() for node in bi_partite_graph.vs}

    indegree_red_nodes = {}
    indegree_blue_nodes = {}


    for node in indegree_nodes:
        if mapping_colors[node] == "red":
            indegree_red_nodes[node] = indegree_nodes[node]
        elif mapping_colors[node] == "blue":
            indegree_blue_nodes[node] = indegree_nodes[node]
            
            
    number_red_nodes = int(5000*s_m)
    
    number_blue_nodes = 5000 - number_red_nodes

    top_nodes_indegree_red = sorted(indegree_red_nodes.items(), key = lambda x: x[1], reverse = True)[:number_red_nodes]
    top_nodes_indegree_blue = sorted(indegree_blue_nodes.items(), key = lambda x: x[1], reverse = True)[:number_blue_nodes]

    top_red_nodes = [node for (node, indegree) in top_nodes_indegree_red]
    top_blue_nodes = [node for (node, indegree) in top_nodes_indegree_blue]

    subgraph = bi_partite_graph.induced_subgraph(top_red_nodes + top_blue_nodes)
    subgraph.summary()

    mapping_colors_subgraph = mapping_colors_function(subgraph)

    file_nodes = "color_nodes_subgraph_" + config + ".tsv"



    with open("../../plot/sample_gephi/" + config + "/" + file_nodes,'w') as f:
        header = ["Id", "color"]
        w = csv.DictWriter(f, delimiter = "\t", fieldnames = header)
        w.writeheader()

        for node in mapping_colors_subgraph:
            w.writerow({"Id": node, "color": mapping_colors_subgraph[node]})



    file_edges = "edge_list_subgraph_" + config + ".tsv"

    with open("../../plot/sample_gephi/" + config + "/" + file_edges, "w") as f:
        header = ["Source", "Target"]
        w = csv.writer(f, delimiter = "\t")
        w.writerow(header)

        for nodes in subgraph.get_edgelist():
            w.writerow(nodes)
```
